```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df.sample(5)
```

# We drop the row number because we don't need this

```python
df.drop('RowNumber',axis='columns',inplace=True)
df.head(5)
```

```python
df.dtypes
```

# Data Visualization Tenure nd Exited

```python
tenure_exited_no=df[df.Exited==0].Tenure
tenure_exited_yes=df[df.Exited==1].Tenure

plt.hist([tenure_exited_yes, tenure_exited_no],color=['green','red'],label=['Exited=1','Exited=0'])
plt.legend()
```

# for loop in each columns and find out all columns unique values

find out only categorical columns not numeric

```python
def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes=='object':
            print( f'{column} : {df[column].unique()}')
```

```python
print_unique_col_values(df)
```

# Convert Yes and No to 1 or 0

```python
df['Gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df.Gender.unique()
```

```python
df.dtypes
```

```python
df1 = pd.get_dummies(data=df, columns=['Surname','Geography'])
df1.columns
```

```python
df1.sample(3)
```

```python
df1.dtypes
```

# Scaling the data

```python
cols_to_scale = ['Tenure','Age','Balance','EstimatedSalary','NumOfProducts','CustomerId']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
df1.head()
```

# Split the data into train nd test model

```python
X = df1.drop('Exited',axis='columns')
y = df1['Exited']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
print('Shape of X_train: ',X_train.shape)
print('Shape of y_train: ',y_train.shape)
print('Shape of X_test: ',X_test.shape)
print('Shape of y_test: ',y_test.shape)
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

 Now we are creating our neural network model 

# Build a model (ANN) in tensorflow/keras

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(2945, input_shape=(2945,), activation='relu'),
    keras.layers.Dense(1480, activation='relu'),
    keras.layers.Dense(740, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

# Predict our model

```python
yp=model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
